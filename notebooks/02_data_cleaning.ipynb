# PCOS Health Analytics Platform

## Notebook 2: Data Cleaning

### Objective

Prepare the main PCOS clinical dataset for analysis by identifying and resolving data quality issues, including missing values, inconsistent column names, unnecessary columns, and data formatting.

### Dataset Used

This notebook uses the **main PCOS clinical dataset** (`PCOS_data_without_infertility.xlsx`), which was explored in Notebook 1.

```python
#Importing Libraries
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

```python
#Loading the datasets
df = pd.read_excel(
    "../data/raw/PCOS_data_without_infertility.xlsx",
    sheet_name="Full_new"
)

df.head()
```

```python
#Creating a Working Copy
df_clean = df.copy()
```

```python
#Removing empty columns
print(df_clean.shape)
df_clean = df_clean.drop(columns=["Unnamed: 44"])
print(df_clean.shape)
```

### Observation

The column `Unnamed: 44` was removed because it contained almost entirely missing values and did not represent a meaningful clinical variable.

```python
#Cleaning Column Names
df_clean.columns = (
    df_clean.columns
    .str.strip()
    .str.replace(r"\s+", " ", regex=True)
)

df_clean.columns.tolist()
```

### Column names
Column names were standardized by removing leading/trailing whitespace and replacing multiple consecutive spaces with a single space. This improves readability and reduces the risk of referencing errors during analysis.

```python
#Handling missing values
missing = df_clean.isnull().sum()

missing[missing > 0]
```

```python
df_clean[df_clean["Fast food (Y/N)"].isnull()]
df_clean[df_clean["Fast food (Y/N)"].isnull()]
```

```python
df_clean["Fast food (Y/N)"].value_counts()
```

```python
df_clean["Fast food (Y/N)"] = df_clean["Fast food (Y/N)"].fillna(
    df_clean["Fast food (Y/N)"].mode()[0]
)
df_clean["Fast food (Y/N)"].isnull().sum()
```

### Handling Missing Values: Fast food (Y/N)

The `Fast food (Y/N)` column contained one missing value. Since this represents only **0.18%** of the dataset (1 out of 541 records), the missing value was imputed using the mode (`1`, representing "Yes"). This approach preserves all patient records while having a negligible impact on the overall data distribution.

```python
df_clean[df_clean["Marraige Status (Yrs)"].isnull()]
```

```python
df_clean["Marraige Status (Yrs)"].median()
```

```python
df_clean["Marraige Status (Yrs)"] = df_clean["Marraige Status (Yrs)"].fillna(
    df_clean["Marraige Status (Yrs)"].median()
)
df_clean.isnull().sum().sum()
```

 ### Handling Missing Values: Marraige Status (Yrs)

The `Marraige Status (Yrs)` column contained one missing value (approximately 0.18% of the dataset). Since this is a numerical variable with only one missing observation, the value was imputed using the median (7 years). Median imputation preserves the dataset while minimizing the influence of outliers.

```python
df_clean.info()
```

```python
#Finding objects column
df_clean.select_dtypes(include="object").columns
df_clean["II beta-HCG(mIU/mL)"].head(10)
df_clean["II beta-HCG(mIU/mL)"].unique()[:20]
```

```python
pd.to_numeric(df_clean["II beta-HCG(mIU/mL)"], errors="coerce").isna().sum()
```

```python
mask = pd.to_numeric(df_clean["II beta-HCG(mIU/mL)"], errors="coerce").isna()
df_clean.loc[mask, ["Patient File No.", "II beta-HCG(mIU/mL)"]]
```

```python
df_clean["II beta-HCG(mIU/mL)"] = df_clean["II beta-HCG(mIU/mL)"].replace(
    "1.99.", "1.99"
)
df_clean["II beta-HCG(mIU/mL)"] = pd.to_numeric(
    df_clean["II beta-HCG(mIU/mL)"]
)
df_clean["II beta-HCG(mIU/mL)"].dtype
```

```python
df_clean.select_dtypes(include="object").columns
```

```python
mask = pd.to_numeric(df_clean["AMH(ng/mL)"], errors="coerce").isna()

df_clean.loc[mask, ["Patient File No.", "AMH(ng/mL)"]]
```

```python
df_clean["AMH(ng/mL)"] = df_clean["AMH(ng/mL)"].replace("a", np.nan)
df_clean["AMH(ng/mL)"] = pd.to_numeric(df_clean["AMH(ng/mL)"])
df_clean["AMH(ng/mL)"] = pd.to_numeric(df_clean["AMH(ng/mL)"])
df_clean["AMH(ng/mL)"] = df_clean["AMH(ng/mL)"].fillna(
    df_clean["AMH(ng/mL)"].median()
)
df_clean.select_dtypes(include="object").columns
df_clean.isnull().sum()
```

```python
df_clean.info()
```

```python
#Saving cleaned dataset
import os

os.makedirs("../data/processed", exist_ok=True)
df_clean.to_csv("../data/processed/pcos_cleaned.csv", index=False)
os.listdir("../data/processed")
```
